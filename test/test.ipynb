```python
import sys
sys.path.append('../')
import cachegenius
import mokemodule

# decorate your module using the CacheGenius tool before any function call
cachegenius(mokemodule)
```

```python
# use the moke module functions
for x in [0] * 5:
    mokemodule.mokescript.d_f_l_oc(x)
    mokemodule.mokescript.nd_f_l_oc(x)
    mokemodule.mokescript.d_s_l_oc(x)
    mokemodule.mokescript.nd_s_l_oc(x)
    mokemodule.mokescript.d_f_h_oc(x)
    mokemodule.mokescript.nd_f_h_oc(x)
    mokemodule.mokescript.d_s_h_oc(x)
    mokemodule.mokescript.nd_s_h_oc(x)

for x in range(5):
    mokemodule.mokescript.d_f_l_noc(x)
    mokemodule.mokescript.nd_f_l_noc(x)
    mokemodule.mokescript.d_s_l_noc(x)
    mokemodule.mokescript.nd_s_l_noc(x)
    mokemodule.mokescript.d_f_h_noc(x)
    mokemodule.mokescript.nd_f_h_noc(x)
    mokemodule.mokescript.d_s_h_noc(x)
    mokemodule.mokescript.nd_s_h_noc(x)
```

```python
# Analyze results
results = cachegenius.report(debug=True)
```

```python
cachegenius.empty()
```

```python
# verify results
for fn_path, is_deterministic, calls_more_than_once, time_gain, cache_size, should_cache in results:
    fn_name = fn_path.split('.')[-1]
    d, s, h, oc = fn_name.split('_')
    assert ((d == 'd') == is_deterministic) or ((is_deterministic is None) and not calls_more_than_once)
    assert (oc == 'oc') == calls_more_than_once
    assert (s == 's') == (time_gain > 0.01) or (not calls_more_than_once and (time_gain == 0.0))
    assert (h == 'h') == (cache_size > 1_000)
    assert (d == 'd' and s == 's' and h == 'l' and oc == 'oc') == should_cache
```
